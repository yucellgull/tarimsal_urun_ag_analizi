```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Tarayıcıyı başlat
driver = webdriver.Chrome()
driver.maximize_window()

# Sayfayı yükle
driver.get("https://klimatikveri.proselector.net/")  # HTML dosyanızın tam yolu

# Tüm verileri tutmak için liste
veriler = []

try:
    # İl/ilçe seçim kutusunu bul
    select_element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "DistrictSelect"))
    )

    # Tüm il seçeneklerini bul
    ilce_options = select_element.find_elements(By.TAG_NAME, "option")

    # Tüm il seçeneklerini tek tek döngüye al
    current_il = ""
    for ilce_option in ilce_options[1:]:  # İlk seçeneği atla (Seç...)
        ilce_adi = ilce_option.text.strip()
        parent_optgroup = ilce_option.find_element(By.XPATH, "ancestor::optgroup")
        il_adi = parent_optgroup.get_attribute("label")
        
        ilce_option.click()
        time.sleep(2)  # Sayfanın yüklenmesini bekle

        # Gerekli verileri çek
        try:
            data = {
                "İl": il_adi,  # Bulunan ilin adı ekleniyor
                "İlçe": ilce_adi,
                "Ortalama Sıcaklık": driver.find_element(By.ID, "sicaklikOrt").text if driver.find_elements(By.ID, "sicaklikOrt") else "Veri yok",
                "Maksimal Sıcaklık": driver.find_element(By.ID, "sicaklikMax").text if driver.find_elements(By.ID, "sicaklikMax") else "Veri yok",
                "Minimal Sıcaklık": driver.find_element(By.ID, "sicaklikMin").text if driver.find_elements(By.ID, "sicaklikMin") else "Veri yok",
                "Ortalama Bağıl Nem": driver.find_element(By.ID, "bagilNemOrt").text if driver.find_elements(By.ID, "bagilNemOrt") else "Veri yok",
                "Yağış Miktarı": driver.find_element(By.ID, "yagisMiktariOrt").text if driver.find_elements(By.ID, "yagisMiktariOrt") else "Veri yok",
                "Yağışlı Gün Sayısı": driver.find_element(By.ID, "yagisliGunSayisi").text if driver.find_elements(By.ID, "yagisliGunSayisi") else "Veri yok",
                "Güneşlenme Saati": driver.find_element(By.ID, "guneslenmeSaati").text if driver.find_elements(By.ID, "guneslenmeSaati") else "Veri yok",
                "Atmosfer Basıncı": driver.find_element(By.ID, "atmosferBasinci").text if driver.find_elements(By.ID, "atmosferBasinci") else "Veri yok",
                "Rakım": driver.find_element(By.ID, "rakim").text if driver.find_elements(By.ID, "rakim") else "Veri yok",
            }
            print(f"{il_adi} - {ilce_adi} verisi çekildi.")
            veriler.append(data)
        except Exception as e:
            print(f"Veri çekme hatası: {il_adi} - {ilce_adi} - {e}")

finally:
    driver.quit()

# Veriyi pandas DataFrame'e dönüştür ve kaydet
df = pd.DataFrame(veriler)
df.to_csv("iklim_verileriAdana.csv", index=False)
print("Veri seti başarıyla kaydedildi: iklim_verileri.xlsx")
```
